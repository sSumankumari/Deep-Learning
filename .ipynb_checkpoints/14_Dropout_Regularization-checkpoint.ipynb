# Dropout Regularization to help with Overfitting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./sonar_dataset.csv", header=None)
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis='columns')
y = df[60]
```

```python
X
```

```python
y
```

### Using one hot encoding

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5)
```

```python
# Convert boolean values to integer
y = y[y.select_dtypes(include=['bool']).columns].astype(int)
# R --> 1 and M --> 0
y
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred[:10]
```

```python
# Round the values to nearest integer i.e, 0 to 1
y_pred = np.round(y_pred)
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

### Adding Dropout layer

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred1 = modeld.predict(X_test).reshape(-1)
y_pred1[:10]
```

```python
# Round the values to nearest integer i.e, 0 to 1
y_pred1 = np.round(y_pred1)
y_pred1[:10]
```

```python
y_test[:10]
```

```python
print(classification_report(y_test, y_pred1))
```
